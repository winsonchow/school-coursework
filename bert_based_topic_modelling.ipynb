## 1.4 Evaluation methodology

To assess the quality and effectiveness of the developed topic modelling system for the 20 Newsgroups dataset, a comprehensive evaluation methodology will be employed. This methodology incorporates a combination of quantitative metrics and qualitative analysis, ensuring a thorough assessment of the system's performance. The following methods and metrics will be used, along with their convincing rationale:

#### 1. Accuracy
Accuracy is a performance metric that measures the overall correctness of a model's predictions. It calculates the ratio of correct predictions to the total number of predictions made, providing an indication of the model's ability to classify instances correctly. However, accuracy can be misleading when dealing with imbalanced datasets, where the number of instances in different classes is significantly different.

#### 2. Precision
Precision is a metric that quantifies the proportion of true positive predictions out of the total predicted positives. It focuses on the correctness of positive predictions, aiming to minimize false positives. A high precision score indicates a low rate of falsely predicted positive instances.

#### 3. Recall
Recall is a measure of a model's ability to identify all of the positive instances in a dataset. It is calculated as the ratio of true positives to the sum of true positives and false negatives. A high recall score indicates that the model is good at identifying positive instances, even if it also produces some false positives.

#### 4. F1-score
The F1 score is a measure of a model's performance that takes into account both precision and recall. It is the harmonic mean of precision and recall, which means that it gives equal weight to both metrics. The F1 score is useful when there is a trade-off between precision and recall. For example, a model that is tuned for high precision may have low recall, and vice versa. The F1 score can be used to find a balance between these two metrics, so that the model is not too sensitive to either one.

#### 5. Support: 
Support refers to the number of instances belonging to a particular class in the dataset. It indicates the frequency of occurrence of a class and provides context for understanding the performance of a model on that class. Support can be used to identify classes with low representation, which may affect the reliability of evaluation metrics for those classes.

#### 6. Comparison with Baselines:
To establish a baseline for comparison, the developed topic modelling system will be compared against existing state-of-the-art approaches, such as BERT modelling. This allows for a comparative analysis of the system's performance. The comparison will provide evidence of the system's effectiveness and highlight its potential contributions in advancing topic modelling techniques.

By employing these evaluation methods and metrics, the evaluation methodology ensures a comprehensive and rigorous assessment of the developed topic modelling system. The combination of quantitative metrics, human evaluation, comparison with baselines, and sensitivity analysis provides a holistic understanding of the system's performance, highlighting its strengths, weaknesses, and potential for improvement. The convincing rationale behind the chosen evaluation methods strengthens the validity and reliability of the evaluation process.

## 2.2 Pre-processing

### 2.2.1 Library imports

```python
# Importing the re module for regular expression operations
import re

# Importing the nltk library for natural language processing tasks
import nltk

# Importing the stopwords corpus from the nltk library
from nltk.corpus import stopwords

# Importing the WordNetLemmatizer class from the nltk.stem module
from nltk.stem import WordNetLemmatizer

# Importing the word_tokenize function from the nltk.tokenize module
from nltk.tokenize import word_tokenize
```

### 2.2.2 Defining pre-processing functions

#### Text cleaning
By removing non-alphabetic characters, numbers, and special symbols, we eliminate noise and irrelevant information, allowing the focus to be on the essential textual content. Converting the text to lowercase ensures consistency in word representations and avoids duplication based on case sensitivity. Additionally, removing stop words, which are commonly occurring and less informative words, helps to refine the text by eliminating noise and reducing dimensionality, ultimately improving the quality of subsequent text analysis and topic modelling results.

```python
def clean_text(text):
    """
    Cleans a text string by removing non-alphabetic characters, converting the text to lowercase,
    and removing stop words.

    Args:
        text (str): The text string to be cleaned.

    Returns:
        str: The cleaned text string.
    """

    # Remove non-alphabetic characters
    text = re.sub(r'[^a-zA-Z]', ' ', text)

    # Convert the text to lowercase
    text = text.lower()

    # Tokenize the text.
    tokens = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Join the tokens back together
    cleaned_text = ' '.join(tokens)

    return cleaned_text
```

#### Lemmatisation
Lemmatisation transforms words to their base or dictionary form, considering their grammatical context and semantic meaning. Unlike stemming, which applies simple rules to strip word endings, lemmatisation produces more accurate and linguistically valid word forms. This is crucial for topic modelling as it helps maintain word integrity, avoids fragmentation of related terms, and improves topic coherence. By capturing the true meanings and relationships between words, lemmatisation enhances the interpretability and accuracy of topic models, making it a preferred technique over stemming in this context.

```python
def lemmatize_text(text):
    """
    Lemmatizes a text string by using the WordNetLemmatizer to find the base form of each word.

    Args:
        text (str): The text string to be lemmatized.

    Returns:
        str: The lemmatized text string.
    """

    # Create a WordNetLemmatizer object
    lemmatizer = WordNetLemmatizer()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Lemmatize each token
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join the tokens back together
    lemmatized_text = ' '.join(lemmatized_tokens)

    return lemmatized_text
```

### 2.2.3 Cleaning the raw 20-Newsgroups dataset

```python
cleaned_documents = []

# Iterate over each document and perform cleaning and lemmatization
for document in documents:
    cleaned_document = clean_text(document)  # Clean the text of the document
    lemmatized_document = lemmatize_text(cleaned_document)  # Lemmatize the cleaned document
    cleaned_documents.append(lemmatized_document)  # Add the lemmatized document to the list

document_index = 0

# Print the original document
print("Original Document:")
print(documents[document_index])

# Print the cleaned document
print("\nCleaned Document:")
print(cleaned_documents[document_index])
```

## 2.3 Establising a baseline

### 2.3.1 Baseline description

The Naive Bayes classifier is a popular and widely used algorithm for topic modeling tasks. It is based on the principles of Bayesian probability and assumes that features (words in this case) are conditionally independent given the class label. Despite its simplistic assumptions, Naive Bayes has shown to be effective in various text classification tasks, including topic modeling.

### 2.3.2 Justifiction

Naive Bayes is often used as a baseline for classification tasks due to its simplicity, efficiency, and effectiveness in many real-world scenarios. Its assumption of conditional independence among features makes it computationally lightweight and fast to train. Despite its simplifying assumptions, Naive Bayes can provide surprisingly competitive performance, especially when the dataset is large and high-dimensional. Comparing more sophisticated models against Naive Bayes allows us to gauge the improvement in performance, and assess the added complexity and computational costs of advanced algorithms.

### 2.3.3 Library imports

```python
# Importing the required libraries for the baseline model

# For converting text data into TF-IDF vectors
from sklearn.feature_extraction.text import TfidfVectorizer

# Naive Bayes classifier
from sklearn.naive_bayes import MultinomialNB

# Evaluation metrics for the model
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix

# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
```

### 2.3.4 Implementing the baseline

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(cleaned_documents, labels, test_size=0.2, random_state=42)

# Preprocess the text data
vectorizer = TfidfVectorizer()
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)

# Initialize and train the Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(x_train, y_train)

# Make predictions on the test set
y_pred = nb_classifier.predict(x_test)
```

### 2.3.5 Baseline evaluation

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report = classification_report(y_test, y_pred, target_names=categories)
```

```python
# Print the evaluation metrics
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("\nClassification Report:\n", classification_report)
```

```python
# Compute the confusion matrix
baseline_cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(15, 25))
sns.heatmap(baseline_cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.xticks(ticks=range(len(categories)), labels=categories, rotation=90)
plt.yticks(ticks=range(len(categories)), labels=categories)
plt.show()
```

## 2.4 Establising a classifier

### 2.4.1 Classifier description

BERT (Bidirectional Encoder Representations from Transformers) is a state-of-the-art natural language processing model developed by Google. It is based on a transformer architecture that allows it to capture the contextual relationships and meanings of words in a sentence by considering the surrounding words from both left and right directions.

BERT utilizes unsupervised pre-training on a large corpus of text data, followed by fine-tuning on a specific downstream task. During pre-training, BERT learns to predict missing words within sentences and to understand the relationships between sentences. This process enables BERT to acquire rich contextual representations of words, which can be beneficial for a variety of NLP tasks, including text classification, named entity recognition, and sentiment analysis.

### 2.4.2 Justification

BERT is a valuable tool for topic modelling due to its ability to capture fine-grained contextual information and semantic relationships within text. By leveraging its pre-trained language representation capabilities, BERT can generate rich and meaningful representations for each word in a document, taking into account its surrounding context. This contextual understanding allows BERT to better grasp the intricate nuances and subtle differences in meaning, enabling more accurate topic classification.

Furthermore, BERT's bidirectional nature ensures that it can effectively capture dependencies between words and capture the global context of the entire document, rather than relying solely on local information. This is particularly beneficial for topic modelling, as it helps identify complex topic structures and detect subtle topic shifts within a document or across a corpus.

BERT's versatility also extends to its fine-tuning capabilities. Researchers can leverage BERT's pre-trained weights and adapt them to specific topic classification tasks by fine-tuning on labeled data. This flexibility allows BERT to adapt to different domains and topics, making it suitable for topic modelling in a wide range of applications.

### 2.4.3 Library imports

```python
# Importing the ktrain library
import ktrain

# Importing the text module from ktrain library
from ktrain import text
```

### 2.4.4 Implementing the classifier

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(cleaned_documents, labels, test_size=0.2, random_state=42)

# Load the dataset and preprocess the text data using BERT embeddings
(x_train, y_train), (x_test, y_test), preproc = text.texts_from_array(x_train=x_train, y_train=y_train,
                                                                       x_test=x_test, y_test=y_test,
                                                                       class_names=categories,
                                                                       preprocess_mode='bert',
                                                                       maxlen=350, 
                                                                       max_features=35000)
```

```python
# Create a BERT-based text classification model
model = text.text_classifier('bert', train_data=(x_train, y_train), preproc=preproc)

# Initialize the learner and set the batch size data
learner = ktrain.get_learner(model, train_data=(x_train, y_train), batch_size=10)
```

```python
# Train the model using the one-cycle learning rate policy
learner.fit_onecycle(2e-5, 3)
```

### 2.4.5 Classifier evaluation

```python
# Validate the model on the testing set
learner.validate(val_data=(x_test, y_test), class_names=categories)
```
